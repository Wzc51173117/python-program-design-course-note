### Matplotlib

#### 绘制简单图形

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成数据
x = np.arange(0, 6, 0.1) # 以0.1为单位，生成0到6的数据
y = np.sin(x)

# 绘制图形
# 设置图像大小（宽度, 高度），单位是英寸
plt.figure(figsize=(3.5, 2.5), dpi=100) # dpi 越高，图像越清晰（但文件越大）, dpi默认为100.  
plt.plot(x, y, label="sin")
plt.title('sin')        # 标题
plt.legend() # 如果你在绘图时已经用 label 参数指定了标签，直接调用 plt.legend() 即可显示图例
plt.show()
```

#### pyplot的功能

```python
# 生成数据
x = np.arange(0, 6, 0.1) # 以0.1为单位，生成0到6的数据
y1 = np.sin(x)
y2 = np.cos(x)

# 绘制图形
plt.figure(figsize=(3.5, 2.5), dpi=100)
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos") # 用虚线绘制
plt.xlabel("x")         # x轴标签
plt.ylabel("y")         # y轴标签

plt.title('sin & cos')  # 标题
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()
plt.show()
```

```python
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].hist(data[0])                   # 子图 (0,0): data[0] 的直方图
axs[1, 0].scatter(data[0], data[1])       # 子图 (1,0): data[0] vs data[1] 的散点图
axs[0, 1].plot(data[0], data[1])          # 子图 (0,1): data[0] vs data[1] 的折线图
axs[1, 1].hist2d(data[0], data[1])        # 子图 (1,1): data[0] 和 data[1] 的二维密度直方图

plt.show()
```

#### 显示图像

```python
from matplotlib.image import imread

img = imread('lena.jpg')       # 读入图像（设定合适的路径）
plt.imshow(img)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 12))
fig.add_subplot(1, 3, 1)  # 创建 1 行 3 列的子图网格，选择第 1 个子图
plt.imshow(img[:,:,0], cmap='gray')  # 提取彩色图像的红色通道（假设 img 是形状为 (H,W,3) 的 NumPy 数组）
# cmap='gray'使用灰度颜色映射显示单通道图像
plt.title('R')

fig.add_subplot(1, 3, 2)
plt.imshow(img[:,:,1], cmap='gray')
plt.title('G')

fig.add_subplot(1, 3, 3)
plt.imshow(img[:,:,2], cmap='gray')
plt.title('B')

plt.show()
```
